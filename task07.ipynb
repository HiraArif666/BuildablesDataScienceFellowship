### **Statistical Analysis and Probability Distributions with the Iris Dataset**

```python
# ==============================================================================
# Statistical Analysis and Probability Distributions with the Iris Dataset
# ==============================================================================

# Import necessary libraries for data handling, plotting, and statistical analysis
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset from scikit-learn
iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['species'])

# Convert 'species' column to actual species names for better readability
df['species'] = df['species'].astype(int).map(dict(enumerate(iris['target_names'])))

# Print the first few rows to verify the data is loaded correctly
print("Initial Data Snapshot:")
print(df.head())
print("-" * 50)


# ==============================================================================
# ðŸ“Š Task 1: Descriptive Statistics - Central Tendencies
# ==============================================================================

# ------------------------------------------------------------------------------
# Central Tendencies: Mean, Median, and Mode of Sepal Length
# ------------------------------------------------------------------------------
print("1. Central Tendencies for Sepal Length")
sepal_length_mean = df['sepal length (cm)'].mean()
sepal_length_median = df['sepal length (cm)'].median()
sepal_length_mode = df['sepal length (cm)'].mode()

print(f"Mean of Sepal Length: {sepal_length_mean:.2f} cm")
print(f"Median of Sepal Length: {sepal_length_median:.2f} cm")
print(f"Mode of Sepal Length: {sepal_length_mode[0]:.2f} cm")

print("-" * 50)
```

 **Interpretation:
The mean, median, and mode are all very close, suggesting that the distribution
 of sepal lengths is fairly symmetrical and does not have extreme outliers.
 The mean gives the average length, the median is the middle value, and the
 mode is the most frequent value, and their proximity here indicates a
 well-behaved, non-skewed distribution.**

```python
# ==============================================================================
# ðŸ“Š Task 2: Descriptive Statistics - Dispersion Measures
# ==============================================================================

# ------------------------------------------------------------------------------
# Dispersion Measures: Variance and Standard Deviation of Petal Width
# ------------------------------------------------------------------------------
print("2. Dispersion Measures for Petal Width")
petal_width_variance = df['petal width (cm)'].var()
petal_width_std = df['petal width (cm)'].std()

print(f"Variance of Petal Width: {petal_width_variance:.4f}")
print(f"Standard Deviation of Petal Width: {petal_width_std:.4f} cm")

print("-" * 50)
```

**Interpretation:
The variance measures how far the data points are spread out from the average.
 The standard deviation, which is the square root of the variance, is
 more easily interpretable because it is in the same units as the original
 data (cm). A standard deviation of 0.76 is relatively large compared to the
 mean petal width, suggesting that petal widths are quite spread out. This
 makes sense given that the dataset contains three different species with
 distinct petal width ranges.**

```python
# ==============================================================================
# ðŸ“Š Task 3: Descriptive Statistics - Summary Statistics
# ==============================================================================

# ------------------------------------------------------------------------------
# Summary Statistics for all Numerical Features
# ------------------------------------------------------------------------------
print("3. Summary Statistics Table")
summary_stats = pd.DataFrame({
    'Mean': df.mean(numeric_only=True),
    'Median': df.median(numeric_only=True),
    'Variance': df.var(numeric_only=True),
    'Standard Deviation': df.std(numeric_only=True)
})
print(summary_stats)

print("-" * 50)
```

**Interpretation:
 This table provides a concise overview of all numerical features.
  Sepal length and width have relatively low variance, meaning they are
   fairly consistent across the dataset.
  Petal length and width show significantly higher variance and standard
   deviation, which is expected as these features are the primary
   differentiating characteristics between the different Iris species.**

```python
# ==============================================================================
# ðŸ“ˆ Task 4: Probability & Distributions - Probability Distribution
# ==============================================================================

# ------------------------------------------------------------------------------
# Probability Distribution of Sepal Length
# ------------------------------------------------------------------------------
print("4. Probability Distribution of Sepal Length (Empirical)")
# Using value_counts() with normalize=True to get the probability of each unique value
sepal_length_distribution = df['sepal length (cm)'].value_counts(normalize=True).sort_index()
print(sepal_length_distribution)

print("-" * 50)
```

**Interpretation:
 This table shows the empirical probability for each observed sepal length.
 For example, a sepal length of 5.0 cm has a probability of 0.10, meaning
 that approximately 10% of the flowers in the dataset have that sepal length.
 The sum of all these probabilities equals 1.**

```python
# ==============================================================================
# ðŸ“ˆ Task 5: Probability & Distributions - Histogram Visualization
# ==============================================================================

# ------------------------------------------------------------------------------
# Histogram Visualization of Sepal Length Probability
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
print("5. Histogram of Sepal Length")
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal length (cm)'], bins=15, kde=True, stat='probability')
plt.title('Probability Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Probability')
plt.show()

print("-" * 50)
```

**Interpretation:
 The histogram visually represents the probability distribution. The shape
 shows that the most probable sepal lengths are clustered around 5.0 cm
 and 6.0 cm, with fewer instances at the extreme ends. The bell-like
 curve of the Kernel Density Estimate (KDE) suggests a roughly normal
 distribution.**

```python
# ==============================================================================
# ðŸ“ˆ Task 6: Probability & Distributions - Cumulative Distribution Function (CDF)
# ==============================================================================

# ------------------------------------------------------------------------------
# Cumulative Distribution Function (CDF) for Petal Length
# ------------------------------------------------------------------------------
print("6. Cumulative Distribution Function (CDF) for Petal Length")
plt.figure(figsize=(8, 6))
sns.ecdfplot(data=df, x='petal length (cm)')
plt.title('Cumulative Distribution Function (CDF) of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Cumulative Probability')
plt.show()

print("-" * 50)
```

**Interpretation:
 The CDF shows the cumulative probability of a petal length being less than or
 equal to a certain value. For example, by looking at the plot, we can see
 that the probability of a petal length being 2 cm or less is close to 0.4.
 The steep jumps in the plot indicate a high concentration of data points at
 those particular petal lengths, which is due to the three distinct species.**

```python
# ==============================================================================
# ðŸ“ˆ Task 7: Probability & Distributions - Probability Density Function (PDF)
# ==============================================================================

# ------------------------------------------------------------------------------
# Probability Density Function (PDF) for Sepal Width
# ------------------------------------------------------------------------------
print("7. Probability Density Function (PDF) for Sepal Width")
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='sepal width (cm)', fill=True, color='skyblue')
plt.title('Probability Density Function (PDF) of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()

print("-" * 50)
```

**Interpretation:
 The PDF shows the relative likelihood of a given value occurring. The peaks
 in the plot indicate regions where the sepal width is most common. This plot
 is particularly useful for visualizing the distribution's shape and is a
 smoothed version of a histogram.**

```python
# ==============================================================================
# ðŸ“ˆ Task 8: Probability & Distributions - Probability Estimation
# ==============================================================================

# ------------------------------------------------------------------------------
# Probability Estimation: P(Petal Length > 4.5 cm)
# ------------------------------------------------------------------------------
print("8. Probability Estimation")
petal_length_threshold = 4.5
probability_greater_than_threshold = (df['petal length (cm)'] > petal_length_threshold).mean()

print(f"Probability of Petal Length > {petal_length_threshold} cm: {probability_greater_than_threshold:.2%}")

print("-" * 50)
```

**Interpretation:
 This is a direct calculation of the empirical probability from the dataset.
 The result of 58.00% means that if you were to randomly select one flower
 from this dataset, there is a 58% chance that its petal length would be
 greater than 4.5 cm.**

```python
# ==============================================================================
# ðŸ”¬ Task 9: Inferential Statistics - Hypothesis Testing
# ==============================================================================

# ------------------------------------------------------------------------------
# Hypothesis Testing (t-test)
# ------------------------------------------------------------------------------
from scipy import stats
print("9. Hypothesis Testing: T-test for Petal Length (Setosa vs. Versicolor)")
# Separate the data for the two species
setosa_petal_lengths = df[df['species'] == 'setosa']['petal length (cm)']
versicolor_petal_lengths = df[df['species'] == 'versicolor']['petal length (cm)']

# Perform the independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(setosa_petal_lengths, versicolor_petal_lengths, equal_var=False)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.10f}")

print("-" * 50)
```

**Interpretation:
 Our null hypothesis (Hâ‚€) is that there is no significant difference in the
 mean petal length between Setosa and Versicolor species.
 The very small p-value (much less than the common significance level of 0.05)
 provides strong evidence to reject the null hypothesis.
 Conclusion: There is a statistically significant difference in the mean
 petal length between Setosa and Versicolor species.**
