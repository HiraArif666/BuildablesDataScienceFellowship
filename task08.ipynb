## **Introduction to NumPy and Pandas for Data Manipulation**

### **1. NumPy Array Creation**

We can create a 1D NumPy array from a list and a 2D array from a list of lists. Basic operations like addition and multiplication are performed element-wise. Slicing works similarly to Python lists.

```python
import numpy as np

# Create 1D array
arr1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1d)

# Create 2D array
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2d)

# Basic operations (element-wise)
print("Addition:", arr1d + 5)
print("Multiplication:\n", arr2d * 2)

# Slicing
print("Slicing 1D array:", arr1d[1:4])
print("Slicing 2D array:", arr2d[0, 1])
```

### **2. Array Statistics**

NumPy provides functions to quickly compute descriptive statistics on arrays.

```python
# NumPy statistical functions
arr = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Index of max value:", np.argmax(arr))
```

### **3. Pandas Series and Dataframe**

A Series is a 1D labeled array, and a DataFrame is a 2D labeled data structure with columns of potentially different data types.

```python
import pandas as pd

# Create a Series from a list
s = pd.Series([10, 20, 30, 40], name='Ages')
print("Pandas Series:\n", s)

# Create a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['NY', 'LA', 'Chicago']}
df = pd.DataFrame(data)
print("\nPandas DataFrame:\n", df)
```

### **5. Data Cleaning**

Data cleaning is essential for any data analysis workflow, and handling missing values is a key part of this process. Missing values are often represented as NaN (Not a Number) in pandas. You can handle them with two primary methods: fillna() and dropna().

```python
# A good practice is to first check the number of null values per column
print("Number of missing values per column (before filling):")
print(df_windows.isnull().sum())
```

```python
import pandas as pd
import numpy as np

# Create a list of all strings that represent missing values
missing_values = ['NULL', 'null', 'none', 'no' ,'Not in Education' ,'06-Oct' ,'.' ,'-', '----' ,'na' ,'n/a' ,'N' ,'nil' ,'no institution' ,'No Name' ,'No school' ,'Nothing','Not in school','Not in education','Not Applicable']
# Now, use fillna() to replace all np.nan with 'Not Specified'
df_windows.fillna('Not Specified', inplace=True)

# Check the DataFrame after cleaning
print("After cleaning:")
print(df_windows)
print("-" * 30)

```

### **7. Sorting and Filtering**

Filtering selects a subset of rows that meet specific conditions. 

```python
# Find learners from Pakistan who have a 'Bs' degree
pakistan_bachelor_learners = df_windows[(df_windows['country'] == 'Pakistan') & (df_windows['major'] == 'Bs')]
print("\nLearners from Pakistan with a B.S. degree:\n", pakistan_bachelor_learners.head())
```

Sorting reorders the DataFrame based on column values.

```python
#sorting countries in ascending order
df_sorted_by_country = df_windows.sort_values(by='country')
print("\nDataFrame sorted by Country (ascending):\n", df_sorted_by_country['country'].head(5))
```

### **8. Groupby and Aggregation**

The .groupby() method is used to group data by a specific column and apply an aggregate function.

```python
# Number of learners per country
learners_per_country = df_windows.groupby('country')['learner_id'].count()
print("Number of learners per country:\n", learners_per_country)
```

```python
# Count different degrees per country
degree_counts_by_country = df_windows.groupby(['country', 'degree'])['learner_id'].count()
print("\nNumber of different degrees per country:\n", degree_counts_by_country.head())
```

### **9. Merging and Joining**

pd.merge(): Use this to combine two DataFrames based on a common key, similar to a database join.

```python
import pandas as pd

# Load the new CSV file, assuming it's in the same directory
df_opportunity = pd.read_csv(r'C:\Users\ilaib\OneDrive\Desktop\Excelerate datasets\LearnerOpportunity_Raw.csv')

# Display the first few rows to inspect its structure
print("LearnerOpportunity_Raw DataFrame:\n", df_opportunity.head())
```

```python
# Merge df_windows (left) with df_opportunity (right)
merged_df = pd.merge(df_windows, df_opportunity, left_on='learner_id', right_on='enrollment_id', how='left')

# Display the first few rows of the merged DataFrame to verify
print("\nMerged DataFrame:\n", merged_df.head())
```

```python
# Drop the redundant 'enrollment_id' column
merged_df.drop('enrollment_id', axis=1, inplace=True)

print("\nMerged DataFrame after dropping redundant column:\n", merged_df.head())
```

### **1. Line Plot Basics**

```python
import matplotlib.pyplot as plt

# Sample data representing a trend over time
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

# Create a line plot
plt.plot(x, y)

# Add title and labels
plt.title("Simple Line Plot")
plt.xlabel("Time")
plt.ylabel("Values")

# Display the plot
plt.show()
```

### **2. Bar Chart Frequencies**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the number of learners in each country and get the top 10
top_10_countries = df_windows['country'].value_counts().head(10)

print("Top 10 countries by learner count:")
print(top_10_countries)
```

```python
# Extract the country names and their counts
countries = top_10_countries.index
counts = top_10_countries.values

# Create the bar plot
plt.figure(figsize=(10, 6)) # Adjust figure size for better readability
plt.bar(countries, counts, color='darkgreen')

# Add labels and a title
plt.title("Top 10 Countries by Learner Count", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Learners", fontsize=12)

# Rotate the country labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent labels from being cut off
plt.tight_layout()

# Display the plot
plt.show()
```

### **6. Correlation Heatmap (Dummy Data)**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame with numerical columns
data = {'A': [10, 20, 30, 40, 50],
        'B': [15, 25, 35, 45, 55],
        'C': [50, 40, 30, 20, 10]}
df_sample = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df_sample.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

plt.title('Correlation Heatmap of Sample Data')
plt.show()
```

### **9. Violin Plot Distribution (Dummy Data)**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame with a numerical and a categorical column
data = {'category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
        'value': [10, 12, 11, 25, 28, 26, 40, 42, 41]}
df_sample = pd.DataFrame(data)

# Create a violin plot
sns.violinplot(x='category', y='value', data=df_sample)

# Add a title and labels
plt.title("Violin Plot of Value Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

### **10. Hybrid Visualization (Dummy Data)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a sample DataFrame with a numerical column
np.random.seed(42) # for reproducibility
data = {'value': np.random.normal(loc=50, scale=10, size=200)}
df_sample = pd.DataFrame(data)

# Create a figure and axes
plt.figure(figsize=(10, 6))

# Use Matplotlib to create a histogram
plt.hist(df_sample['value'], bins=20, color='skyblue', edgecolor='black', alpha=0.7, label='Histogram')

# Use Seaborn to overlay a KDE plot on the same axes
sns.kdeplot(df_sample['value'], color='red', linestyle='--', linewidth=2, label='KDE')

# Add a title and labels
plt.title("Hybrid Visualization: Histogram with KDE Overlay", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()

# Display the plot
plt.show()
```
