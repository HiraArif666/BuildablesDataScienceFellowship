**Task 01: Load a built-in dataset (e.g., the  from Scikit-Learn) and display the first 7 rows.**

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
 
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
 
print("First 7 rows of the Iris dataset:")
print(iris_df.head(7))
print("-" * 50)
```

**Task 02: Write a function that accepts a dataset and returns the shape information (number
of rows and columns).**

```python
def get_shape(dataset):
    """Returns the shape of a given dataset."""
    return dataset.shape
 
print(f"Shape of the dataset: {get_shape(iris_df)}")
print("-" * 50)
```

**Task 03: Divide the dataset into training and testing subsets using a 70/30 ratio.**

```python
X = iris_df.drop('species', axis=1)
y = iris_df['species']
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
 
print("Shape of training data:")
print(f"Features: {X_train.shape}, Target: {y_train.shape}")
print("\nShape of testing data:")
print(f"Features: {X_test.shape}, Target: {y_test.shape}")
print("-" * 50)
```

**Task 07 : Define a function that takes a list of values and outputs a summary dictionary with
count, mean, min, max, and standard deviation.**

```python
def get_summary_stats(values):
    """Outputs a summary dictionary with key statistics for a list of values."""
    values = np.array(values)
    summary = {
        'count': len(values),
        'mean': np.mean(values),
        'min': np.min(values),
        'max': np.max(values),
        'std_dev': np.std(values)
    }
    return summary
 
print("Summary statistics for the list of numbers:")
print(get_summary_stats(numbers_list))
print("-" * 50)
```

**Task 10 : Build and train a simple linear regression model using Scikit-Learn on any dataset,
and print both the coefficients and intercept.**

```python
# Select two numerical features for linear regression
X_lin_reg = iris_df[['sepal length (cm)']]
y_lin_reg = iris_df['sepal width (cm)']
 
# Create and train the model
lin_reg = LinearRegression()
lin_reg.fit(X_lin_reg, y_lin_reg)
 
print("Linear Regression Model:")
print(f"Coefficient: {lin_reg.coef_[0]}")
print(f"Intercept: {lin_reg.intercept_}")
print("-" * 50)
```

**Task 11 : Normalize one numerical column in the dataset using Min-Max scaling and print the
first 10 transformed values.**

```python
scaler = MinMaxScaler()
iris_df['scaled_petal_length'] = scaler.fit_transform(iris_df[['petal length (cm)']])
 
print("First 10 normalized 'petal length (cm)' values:")
print(iris_df['scaled_petal_length'].head(10))
print("-" * 50)
```

**Task 12 : Group the dataset by one categorical feature and calculate the mean of a numerical

**Task 13 : Plot a scatterplot between two numerical features from the dataset to observe their

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=iris_df)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
print("-" * 50)
```

**Task 14 : Implement a correlation heatmap for all numerical features in the dataset using
Seaborn or Matplotlib.**

```python
numerical_features = iris_df.select_dtypes(include=np.number).drop(columns=['species'], errors='ignore')
correlation_matrix = numerical_features.corr()
 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
print("-" * 50)
```

**Task 15 : Save a cleaned version of your dataset (with missing values handled) into a new
CSV file.**

```python
# Create a dummy DataFrame with missing values
dummy_df_with_na = iris_df.copy()
dummy_df_with_na.loc[0, 'sepal length (cm)'] = np.nan
dummy_df_with_na.loc[10, 'sepal length (cm)'] = np.nan
 
print("Original DataFrame with missing values:")
print(dummy_df_with_na.head(12))
 
# Handle missing values (fill with mean)
cleaned_df = dummy_df_with_na.copy()
mean_sepal_length = cleaned_df['sepal length (cm)'].mean()
cleaned_df['sepal length (cm)'].fillna(mean_sepal_length, inplace=True)
 
print("\nCleaned DataFrame with missing values filled:")
print(cleaned_df.head(12))
 
# Save the cleaned dataset to a new CSV file
cleaned_df.to_csv('cleaned_iris_dataset.csv', index=False)
print("\nCleaned dataset saved to 'cleaned_iris_dataset.csv'.")
print("-" * 50)
```
